# LLCBench Quickstart

This notebook walks through the **end-to-end** flow:

1. Inspect the published results that ship with the repository.
2. (Optional) download benchmark LLMs from Hugging Face.
3. (Optional) run a single compressor on a single model.
4. Compute aggregate metrics & plot figures.

````python
import pandas as pd
from pathlib import Path

ROOT = Path('..').resolve()
results = ROOT / 'results'

overall = pd.read_csv(results / 'overall.csv')
overall.head()
````

## 1. Top-3 by AvgCR

````python
overall.sort_values('AvgCR', ascending=False).head(3)
````

## 2. Per-algorithm detail (e.g. ZipNN)

````python
pd.read_csv(results / 'per_algorithm' / 'zipnn.csv')
````

## 3. Run a single algorithm on a single model

```bash
# Compression
bash scripts/tlc/zstd/run.sh ./models/D7.bin /tmp/D7.zstd
# Decompression
bash scripts/tlc/zstd/run.sh -d /tmp/D7.zstd /tmp/D7.restored
```

Each call prints one CSV-formatted metrics line.

## 4. Bar chart of compression ratios

````python
import matplotlib.pyplot as plt
df = overall.sort_values('AvgCR', ascending=False)
ax = df.plot(x='Method', y=['AvgCR','WAvgCR'], kind='bar', figsize=(13,4))
ax.grid(axis='y', linestyle=':', alpha=.4)
ax.set_ylabel('Compression Ratio')
ax.set_title('LLCBench — AvgCR / WAvgCR (higher is better)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
````
